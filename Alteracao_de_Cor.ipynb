```python
import numpy as np
import requests
from io import BytesIO
import imageio.v3 as iio
import matplotlib.pyplot as plt

# 1. Função para converter RGB para tons de cinza
def rgb_to_grayscale(image_rgb):
    height, width, _ = image_rgb.shape
    gray_image = np.zeros((height, width), dtype=np.uint8)

    for i in range(height):
        for j in range(width):
            r, g, b = image_rgb[i, j]
            # Fórmula de luminância perceptiva
            gray = int(0.299 * r + 0.587 * g + 0.114 * b)
            gray_image[i, j] = gray

    return gray_image

# 2. Função para binarizar a imagem
def binarize(gray_image, threshold=127):
    height, width = gray_image.shape
    binary_image = np.zeros((height, width), dtype=np.uint8)

    for i in range(height):
        for j in range(width):
            binary_image[i, j] = 255 if gray_image[i, j] > threshold else 0

    return binary_image

# 3. Baixar a imagem do GitHub
def load_image_from_url(url):
    raw_url = url.replace("github.com", "raw.githubusercontent.com").replace("/blob/", "/")
    response = requests.get(raw_url)
    if response.status_code == 200:
        img = iio.imread(BytesIO(response.content))
        return img
    else:
        raise Exception("Não foi possível baixar a imagem.")

# 4. Executar pipeline
url = "https://github.com/R0YDOUGLAS/Projeto_Alteracao_de_Cor_Python/blob/main/cao.png"
image_rgb = load_image_from_url(url)

gray_image = rgb_to_grayscale(image_rgb)
binary_image = binarize(gray_image)

# 5. Visualizar resultados
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title("Imagem Original")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Imagem em Tons de Cinza")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(binary_image, cmap='gray')
plt.title("Imagem Binarizada (P&B)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
